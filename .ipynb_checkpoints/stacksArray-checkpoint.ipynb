# Stack Implementation in Python
Implementing Stacks in python using List data type

#### Factorial Function Recursive Implementation

```python
def factorial(n):
    
    # Define base case for Recursion
    if n == 0:
        # 0! = 1
        return 1
    else:
        # n! = n * (n-1)!
        return n * factorial(n-1)

# Testing Function
print(factorial(4))
print(factorial(5))
```

#### Implementation

```python
class Stack():
    
    # Constructor method
    def __init__(self):
        self.stack = []
    
    # Method to check if stack is empty 
    def isEmpty(self):
        return self.stack == []
    
    # Method to push items to the stack
    def push(self, item):
        self.stack.append(item)
        
    # Method to pop items from the stack
    def pop2(self):
        # Uses python .pop() method
        return self.stack.pop()
    
    # Method to pop items from the stack
    def pop(self):
        data = self.stack[-1]
        del self.stack[-1]
        return data
    
    # Method to view the top-most stack item
    def peek(self):
        return self.stack[-1]
    
        
```
